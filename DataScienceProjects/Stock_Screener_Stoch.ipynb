```python
def function(tick):

  ticker = yf.Ticker(tick)
  df = ticker.history(period='5y',interval='1d')
  df = df.reset_index(drop=True)
  df = df[["Open", "High", "Low", "Close"]]
  data = np.array([df['Open'], df['High'], df['Low'], df['Close']])

  stoch1, stoch2 = talib.STOCH(data[1], data[2],data[3], fastk_period=20, slowk_period=3, slowd_period=3)
  stoch3, stoch4 = talib.STOCH(data[1], data[2], data[3], fastk_period=50, slowk_period=3, slowd_period=3)
  ema12 = talib.EMA(data[3], 12)

  df['stoch1'],df['stoch2'], df['stoch3'], df['stoch4'] = pd.DataFrame(stoch1), pd.DataFrame(stoch2), pd.DataFrame(stoch3), pd.DataFrame(stoch4)
  df['ema12'] = pd.DataFrame(ema12)
  last = np.array(df.tail(1))


  if (last[0][4] > 50) & (last[0][5] > 50) & (last[0][6] > 50) & (last[0][7] > 50) & (last[0][8] < last[0][3]):
    if (last[0][4] < 65) & (last[0][5] < 65) & (last[0][6] < 65) & (last[0][7] < 65):
      return True
    else:
      return False
  else:
      return False
```
