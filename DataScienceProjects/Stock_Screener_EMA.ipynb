```python
import yfinance as yf
import numpy as np
from plyer import notification
import time

def calculate_ema(data, window):
    return data['Close'].ewm(span=window, adjust=False).mean()

def check_ema_retest(stock_symbol, ema_window, retest_window):
    stock_data = yf.download(stock_symbol, period="1d", interval="5m")
    if len(stock_data) < ema_window + retest_window:
        return False

    ema = calculate_ema(stock_data, ema_window)
    latest_close = stock_data['Close'][-1]
    ema_current = ema[-1]

    if latest_close > ema_current:
        retest_prices = stock_data['Low'][-retest_window:]
        if np.any(retest_prices < ema_current):
            print("YES")

def main():
    stock_symbol = "AAPL"  # Change this to your desired stock symbol
    ema_window = 20
    retest_window = 3

    while True:
        if check_ema_retest(stock_symbol, ema_window, retest_window):
            notification_title = "EMA Retest Buy Notification"
            notification_message = f"{stock_symbol} has retested EMA-{ema_window} and might be a good time to buy."
            notification.notify(title=notification_title, message=notification_message)
        time.sleep(300)  # Wait for 5 minutes before checking again

if __name__ == "__main__":
    main()
```
