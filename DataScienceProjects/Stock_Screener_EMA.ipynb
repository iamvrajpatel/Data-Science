```python
def function(tick):
  ticker = yf.Ticker(tick)
  df = ticker.history(period='5y',interval='1d')
  df = df.reset_index(drop=True)
  df = df[["Open", "High", "Low", "Close"]]
  data = np.array([df['Open'], df['High'], df['Low'], df['Close']])
  ema50 = talib.EMA(data[3], 50)
  df['ema50'] = pd.DataFrame(ema50)
  last = np.array(df.tail(1))
  if ((last[0][3] > last[0][4]) and (last[0][3] < last[0][4] - (last[0][4]*0.04))) or ((last[0][3] < last[0][4]) and (last[0][3] > last[0][4] + (last[0][4]*0.04))):
  # if (last[0][3] < (last[0][5] + (last[0][5]*0.04))) or (last[0][3] > (last[0][5] - (last[0][5]*0.04))):
    return True
  else:
    return False
```
