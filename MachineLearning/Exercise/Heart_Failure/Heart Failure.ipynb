# Handling Text Columns Using Label Encoding 

```python
data2 = data.copy()
data2.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True
)
data2.ST_Slope.replace(
    {
        'Up': 3,
        'Down': 1,
        'Flat': 2
    },
    inplace=True
)
data2.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True
)
data2.head()
```

```python
data3 = pd.get_dummies(data2, drop_first=True)
data3.head()
```

```python
x = data3.drop('HeartDisease', axis='columns')
y = data3.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xscaler = scaler.fit_transform(x)
xscaler
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(xscaler, y, test_size=0.2, random_state=42)
```

## Train Model using Standalone Support Vector Machine and Then using Bagging

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(), x, y, cv = 5)
scores.mean()
```

### Use Bagging With svm 

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(base_estimator=SVC(), n_estimators=100, max_samples=0.8, random_state=42)
scores = cross_val_score(bag_model, x,y, cv=5)
scores.mean()
```

# Train a model using decision tree and then using bagging

```python
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(), x, y, cv=5)
scores.mean()
```

### Use Bagging With svm

```python
bag_model = BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=42),
                              n_estimators=100,
                              max_samples=0.9,
                             oob_score=True,
                             random_state=42
)
scores = cross_val_score(bag_model, x, y, cv = 5)
scores.mean()
```

## Train a Model Using Random Forest Which itself uses bagging 

```python
from sklearn.ensemble import RandomForestClassifier
scores = cross_val_score(RandomForestClassifier(), x, y, cv = 5)
scores.mean()
```
