```python
import pandas as pd
data = pd.read_csv("D:/Data Science/codebasics/titanic2.csv")
data
```

```python
data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin', 'Embarked'], axis='columns', inplace=True)
data
```

```python
target = data.Survived
inputs = data.drop(['Survived'], axis='columns')
target
inputs
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies
```

```python
inputs = pd.concat([inputs, dummies], axis='columns')
inputs
```

```python
inputs = inputs.drop(['Sex'], axis='columns')
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age = inputs.Age.fillna((inputs.Age.mean()))
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(inputs, target, test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
model.predict(xtest[:10])
```

```python
model.predict_proba(xtest[:10])
```
